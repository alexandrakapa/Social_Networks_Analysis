## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

Δεδομένου ότι θέλουμε να υλοποιήσουμε regular lattice θέτουμε $p = 0$.

```python
random.seed(seedno)
n=40 #μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
 
REG = nx.watts_strogatz_graph(n,d,0)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, node_size=80)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=40
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER, node_size=80)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=40
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R) 

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):


#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.


---



Αρχικά υλοποιούμε την συνάρτηση για την εύρεση του συντελεστή ομαδοποίησης κάθε κόμβου:

```python
def cc(G):
  neighbors_of_node=[]
  clustering_coefficient_list=[]
  #find cc for every node
  for node in G.nodes():
    number_of_links = 0
    #find neighbors of each node
    for i in nx.neighbors(G,node):
      neighbors_of_node.append(i)
    # print(neighbors_of_node)
    if (len(neighbors_of_node) > 1):
      for j in neighbors_of_node:
        for k in neighbors_of_node:
          #we need to find an edge between the neighbors
          if G.has_edge(j,k):
            number_of_links = number_of_links + 1
      #2*number_of_links is the number of triangles
      #len(neighbors_of_node) is the degree of the node so len(neighbors_of_node)*(len(neighbors_of_node)-1) is the number of triplets with center the node
      cc = number_of_links/(len(neighbors_of_node)*(len(neighbors_of_node)-1))
      clustering_coefficient_list.append(cc)
      # print(cc)
    else:
      print(0)
    neighbors_of_node=[]
  # print(clustering_coefficient_list)
  return clustering_coefficient_list
```

Στην συνέχεια μας ζητείται να αναπαραστήσουμε την κατανομή αυτού του κόμβου:

```python
def plot_distribution(cc_list):
  y = np.arange(0,40,1)
  x = np.sort(cc_list)
  plt.xlabel("Clustering Coefficient values")
  plt.ylabel("Number of nodes")
  plt.title("Clustering Coefficient diagram")
  plt.hist(x, edgecolor='black', color="indianred")
  plt.show()
```

#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.


---



Ο μέσος συντελεστής διασποράς ορίζεται ως:

$C = \frac{1}{n}\sum_{v{\in}G} c_v$

όπου n o αριθμός των κόμβων




*Σημείωση:*Από εδώ και πέρα θα χρησιμοποιήσουμε την συνάρτηση nx.clustering της networkx για τον υπολογισμό του συντελεστή ομαδοποίησης μιας και αν και δίνει ίδι αποτελέσματα με την παραπάνω συνάρτηση που φτιάξαμε, τα δίνει σε μορφή dict που βολεύει για τις πράξεις.

```python
REG_clust_coef = nx.clustering(REG)
RGER_clust_coef = nx.clustering(RGER)
RGG_clust_coef = nx.clustering(RGG)
SF_clust_coef = nx.clustering(SF)
SW_clust_coef = nx.clustering(SW)
```

```python
def mean_cc(clustering_coefficient):
  return np.mean(list(clustering_coefficient))

def var_cc(clustering_coefficient):
  return np.var(list(clustering_coefficient))
```

#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 


---



Υλοποιήσαμε την παρακάτω συνάρτηση προκειμένου να βρίσκει τους κόμβους με τον μικρότερο και τον μεγαλύτερο συντελεστή ομαδοποίησης. Αν είναι πολλοί κόμβοι με τον ίδιο συντελεστή τότε τους κρατάει όλους σε μορφή λίστας.

```python
def find_min_max(G):
  maxi=0
  mini=10
  coef_list = cc(G)
  maxi = max(coef_list)
  max_index = [index for index, value in enumerate(coef_list) if value == maxi]
  mini = min(coef_list)
  min_index = [index for index, value in enumerate(coef_list) if value == mini]
  return (maxi,max_index,mini,min_index)
```

Για επαλήθευση χρησιμοποιούμε την παρακάτω συνάρτηση που δείχνει με χρήση ενός heatmap χρωματικά τους κόμβους με τον μεγαλύτερο και τον μικρότερο συντελεστή ομαδοποίησης:

```python
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
import networkx as nx

def plot_heatmap(g):
  gc = g.subgraph(max(nx.connected_components(g)))
  lcc = nx.clustering(gc)

  cmap = plt.get_cmap('Dark2')
  norm = plt.Normalize(0, max(lcc.values()))
  node_colors = [cmap(norm(lcc[node])) for node in gc.nodes]
  fig = plt.figure(figsize=(7, 3))
  nx.draw_spring(gc, node_color=node_colors, with_labels=True)
  fig.colorbar(ScalarMappable(cmap=cmap, norm=norm), label='Clustering coefficient', shrink=0.95)
  plt.show()
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):


#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.

---




```python
def compute_mean_path_lengths(G):
  length = dict(nx.all_pairs_shortest_path_length(G))
  lengthlist = 0
  for i in length:
    for j in length[i]:
      lengthlist += length[i][j]
  #We have 40*40 pairs so 1600
  average = lengthlist/1600
  return average
```

```python
def variance_of_path_lengths(G):
  length = list(nx.all_pairs_shortest_path_length(G))
  lengthlist = []
  for i in length:
    j = list(i[1].values())
    lengthlist.append(j)
  return np.var(lengthlist)
```

```python
print('Mean path length of REG topology: ', compute_mean_path_lengths(REG))
print('Variance of path length of REG topology: ', variance_of_path_lengths(REG), '\n')

print('Mean path length of RGER topology: ', compute_mean_path_lengths(RGER))
print('Variance of path length of RGER topology: ', variance_of_path_lengths(RGER), '\n')

print('Mean path length of RGG topology: ', compute_mean_path_lengths(RGG))
print('Variance of path length of RGG topology: ', variance_of_path_lengths(RGG), '\n')

print('Mean path length of SF topology: ', compute_mean_path_lengths(SF))
print('Variance of path length of SF topology: ', variance_of_path_lengths(SF), '\n')

print('Mean path length of SW topology: ', compute_mean_path_lengths(SW))
print('Variance of path length of SW topology: ', variance_of_path_lengths(SW))
```

#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).

---




```python
def find_diam_rad(G):
  ecc=nx.eccentricity(G)
  diam = nx.diameter(G, ecc)
  rad = nx.radius(G,ecc)
  return diam,rad
```

```python
print('The diameter of REG topology: ', find_diam_rad(REG)[0])
print('The radius of REG topology: ', find_diam_rad(REG)[1], '\n')

print('The diameter of RGER topology: ', find_diam_rad(RGER)[0])
print('The radius of RGER topology: ', find_diam_rad(RGER)[1], '\n')

print('The diameter of RGG topology: ', find_diam_rad(RGG)[0])
print('The radius of RGG topology: ', find_diam_rad(RGG)[1], '\n')

print('The diameter of SF topology: ',find_diam_rad(SF)[0])
print('The radius of SF topology: ', find_diam_rad(SF)[1], '\n')

print('The diameter of SW topology: ', find_diam_rad(SW)[0])
print('The radius of SW topology: ', find_diam_rad(SW)[1])
```

Εξαιτίας των hubs βλέπουμε ότι η διάμετρος ισούται με 4, δηλαδή για να φτάσουμε από έναν κόμβο σε έναν άλλο είναι το πολύ 4.
* αντίστοιχα στην τοπολογία SW, όπως αναφέρθηκε παραπάνω οι περισσότεροι κόμβοι δεν είναι γείτονες μεταξύ τους, αλλά οι γείτονες οποιουδήποτε δεδομένου κόμβου είναι πιθανό να είναι γείτονες μεταξύ τους, επομένως έχουμε μικρό μέσο μήκος μονοπατιού.


#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

---



Υλοποιήσαμε την παρακάτω συνάρτηση που βρίσκει ποιός κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με την διάμετρο της τοπολογίας αλλά και με ποιούς κόμβους συνδέεται.

```python
def find_nodes(G):
  length = dict(nx.all_pairs_shortest_path_length(G))
  ecc=nx.eccentricity(G)
  diam = nx.diameter(G, e=ecc)
  # print(diam)
  maximum_nodes_first = 0
  maximum_nodes_second = 0
  the_node = 45
  node_list = []
  for i in length:
    #find the max value to see if diam belongs there
    all_values = length[i].values()
    max_value = max(all_values)
    #if it doesnt exist then we move on
    if (max_value < diam):
         maximum_nodes_first = 0
         continue
    for j in length[i]:
      #if smaller than diam
      if (length[i][j] < diam):
        maximum_nodes_first = 0
      #if the same as diam
      if (length[i][j] == diam):
        #we need to check the node that is connected with max nodes
        maximum_nodes_first += 1
      #the lengths are in order
      if (length[i][j] > diam):
        maximum_nodes_first = 0
        break
    #if more nodes than max then keep this (if there are more we keep the first)
    if (maximum_nodes_first > maximum_nodes_second):
      maximum_nodes_second = maximum_nodes_first
      # print(maximum_nodes_second,i)
      the_node = i
  #if we dont find it return 0 and empty
  if (the_node == 45):
    return 0,[]
  for k in length[the_node]:
     if (length[the_node][k] == diam):
       node_list.append(k)
  # να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
  # με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο. 
  return the_node,node_list
```

```python
print("REG topology:")
REG_node,REG_node_list = find_nodes(REG)
print("The node is",REG_node,"and the nodes it is connected are",REG_node_list,'\n')

print("RGER topology:")
RGER_node,RGER_node_list = find_nodes(RGER)
print("The node is",RGER_node,"and the nodes it is connected are",RGER_node_list,'\n')

print("RGG topology:")
RGG_node,RGG_node_list = find_nodes(RGG)
print("The node is",RGG_node,"and the nodes it is connected are",RGG_node_list,'\n')

print("SF topology:")
SF_node,SF_node_list = find_nodes(SF)
print("The node is",SF_node,"and the nodes it is connected are",SF_node_list,'\n')

print("SW topology:")
SW_node,SW_node_list = find_nodes(SW)
print("The node is",SW_node,"and the nodes it is connected are",SW_node_list,'\n')
```

#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

---


#### ***Περιφέρεια*** του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### ***Κέντρο*** του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  

Υλοποιήσαμε τις δύο παρακάτω συναρτήσεις που υπολογίζουν την περιφέρεια και το κέντρο μιας τοπολογίας.

```python
def find_periphery(G):
    ecc = nx.eccentricity(G) 
    diam = nx.diameter(G, e=ecc)
    circumference = []
    for i in ecc:
      if (ecc[i] == diam):
        circumference.append(i)
    return circumference
```

```python
def find_center(G):
    ecc = nx.eccentricity(G) 
    rad = nx.radius(G, e=ecc)
    central = []
    for i in ecc:
      if (ecc[i] == rad):
        central.append(i)
    return central
```

### Γ1. Κεντρικότητα βαθμού (Degree centrality)

##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.

---

Aρχικά υλοποιούμε την συνάρτηση που βρίσκει τον βαθμό κάθε κόμβου.

```python
def find_degree(G):
  degrees = []
  for i in G.nodes:
      degrees.append((i,G.degree(i)))
  return degrees
```

Και στην συνέχεια υλοποιούμε την συνάρτηση για την αναπαράσταση της κατανομής βαθμών κόμβων.

```python
def plot_degree_distribution(G):
  x = []
  degree_pairs = find_degree(G)
  for i in degree_pairs:
    x.append(i[1])
  y = np.arange(0,40,1)
  plt.xlabel("Degree Centrality values")
  plt.ylabel("Number of nodes")
  plt.title("Degree Centrality diagram")
  plt.hist(x, edgecolor='black', color="indianred")
  plt.show()
```

##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.


---



```python
def avg_var_degree(G):
  degree_pairs = find_degree(G)
  x = []
  for i in degree_pairs:
    x.append(i[1])
  mean = np.mean(x)
  variance = np.var(x)
  return mean,variance
```

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)

---




Γνωρίζουμε ότι η κεντρικότητα εγγύτητας ισούται με το αντίστροφο του μέσου όρου των συντομότερων μονοπατιών το κόμβου προς όλους τους υπολοίπους κόμβους.

##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.


Αρχικά αναπαριστούμε τις κατανομές σε κοινό διάγραμμα:

```python
def find_closeness(G):
  length = dict(nx.all_pairs_shortest_path_length(G))
  closeness_list = []
  for i in length:
    sum = 0
    for j in length[i]:
      sum += length[i][j]
    closeness_list.append(1/(sum/(len(length[i])-1)))
  return closeness_list
```

```python
def plot_closeness_centrality(REG,RGER,RGG,SF,SW):
  closeness_listREG = np.sort(find_closeness(REG))
  closeness_listRGER = np.sort(find_closeness(RGER))
  closeness_listRGG = np.sort(find_closeness(RGG))
  closeness_listSF = np.sort(find_closeness(SF))
  closeness_listSW = np.sort(find_closeness(SW))
  y = np.arange(0,40,1)/40
  plt.figure(figsize=(10,5))
  plt.plot(closeness_listREG,y,label="REG topology",linewidth=3.0)
  plt.plot(closeness_listRGER,y,label="RGER topology",linewidth=3.0)
  plt.plot(closeness_listRGG,y,label="RGG topology",linewidth=3.0)
  plt.plot(closeness_listSF,y,label="SF topology",linewidth=3.0)
  plt.plot(closeness_listSW,y,label="SW topology",linewidth=3.0)
  plt.xlabel("Closeness Centralities")
  plt.ylabel("Nodes")
  plt.legend()
  plt.show
```

```python
plot_closeness_centrality(REG,RGER,RGG,SF,SW)
```

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)

---




##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.

---




Γνωρίζουμε ότι η ενδιάμεση κεντρικότητα κάθε κόμβοθ είναι ο αριθμός των μικρότερων μονοπατιών που περνάνε από αυτόν τον κομβο και ορίζεται ως:

 $C = \sum_{s≠v≠t∈V} \frac{σ_{st}(v)}{σ_{st}}$ 
όπου:
* $σ_{st}(v)$ ο συνολικός αριθμός ελάχιστων μονοπατιών μεταξύ του κόμβου s και του t

* $σ_{st}$ ο αριθμός των ελαχιστών μονοπατιών που περνάνε από τον v

Αρχικά αναπαριστούμε τις κατανομές σε κοινό διάγραμμα:

```python
def find_betweenness(G):
    return nx.betweenness_centrality(G)
```

### Γ4. Kεντρικότητα Katz

---




##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.


Αρχικά αναπαριστούμε τις κατανομές σε κοινό διάγραμμα:

```python
def find_katz(G):
  Gmax = np.max(nx.adjacency_spectrum(G))
  Ga = (1/Gmax.real)-0.01
  Gkc = nx.katz_centrality(G,Ga)
  return Gkc
```

```python
def plot_katz():
  REGkc = find_katz(REG)
  RGERkc = find_katz(RGER)
  RGGkc = find_katz(RGG)
  SFkc = find_katz(SF)
  SWkc = find_katz(SW)

  REGl = []
  RGERl = []
  RGGl = []
  SFl = []
  SWl = []
  for i in range (40):
    REGl.append(REGkc[i])
    RGERl.append(RGERkc[i])
    RGGl.append(RGGkc[i])
    SFl.append(SFkc[i])
    SWl.append(SWkc[i])
  y = np.arange(0,40,1)/40
  plt.figure(figsize=(10,5))
  plt.plot(np.sort(REGl),y,label="REG topology",linewidth=3.0)
  plt.plot(np.sort(RGERl),y,label="RGER topology",linewidth=3.0)
  plt.plot(np.sort(RGGl),y,label="RGG topology",linewidth=3.0)
  plt.plot(np.sort(SFl),y,label="SF topology",linewidth=3.0)
  plt.plot(np.sort(SWl),y,label="SW topology",linewidth=3.0)
  plt.legend()
  plt.xlabel("Node number")
  plt.ylabel("Katz centrality")
  plt.title("Katz centrality distribution")
  plt.show
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    degree_list = nx.degree_centrality(G)
    degree = max(degree_list.values())
    degree_index = [index for index, value in enumerate(degree_list.values()) if value == degree]

    closeness_list = nx.closeness_centrality(G)
    closeness = max(closeness_list.values())
    closeness_index = [index for index, value in enumerate(closeness_list.values()) if value == closeness]

    betweenness_list = nx.betweenness_centrality(G)
    betweenness = max(betweenness_list.values())
    betweenness_index = [index for index, value in enumerate(betweenness_list.values()) if value == betweenness]

    katz_list = find_katz(G)
    katz = max(katz_list.values())
    katz_index = [index for index, value in enumerate(katz_list.values()) if value == katz]
    return degree_index,closeness_index,betweenness_index,katz_index
```

### Ποσοτική ανάλυση

Γνωρίζουμε ότι στην εργασία τους οι Erdős, Rényi χαρακτηρίζουν υπό προϋποθέσεις την συμπεριφορά των G(n, p) γράφων για τις διάφορες τιμές του p και αναφέρουν ότι :
* Αν $p < \frac {( 1 − ε)ln n}{n}$, τότε ο γράφος στο G(n, p) αναμένεται να είναι disconnected.
* Αν $p > \frac {( 1 − ε)ln n}{n}$, τότε ο γράφος στο G(n, p) αναμένεται να είναι connected.

Οπότε η ποσότητα $\frac{ln n}{n}$ συνιστά υπό προϋποθέσεις ένα sharp threshold για την συνδεσιμότητα του γράφου G(n, p). Με βάση το παραπάνω συμπέρασμα, για τον γράφο RG-Gilbert που αναφέραμε έχουμε τα παρακάτω αποτελέσματα σχετικά με τις sharp phase transitions.

```python
 from numpy import log as ln
 y = [p/100 for p in range (1,11,1)]
 #n = 100
 print("For n = 100:\n")
 x = (ln(100))/100
 print("lnn/n is",x,"\n")
 for i in y:
   if i < x:
    print("p is",i,"<",np.round(x,3),"so the graph is disconnected")
   else:
    print("p is",i,">",np.round(x,3),"so the graph is connected")

 #n = 200
 print("For n = 200:\n")
 x = (ln(200))/200
 print("lnn/n is",x,"\n")
 for i in y:
   if i < x:
    print("p is",i,"<",np.round(x,3),"so the graph is disconnected")
   else:
    print("p is",i,">",np.round(x,3),"so the graph is connected")
```

#### Ποιοι ήταν οι 5 κεντρικότεροι χαρακτήρες του πρώτου κύκλου επεισοδίων ως προς το βαθμό, την εγγύτητα και την ενδιαμεσικότητα ? 

---



```python
def find_central_got(G):
    degree_list = nx.degree_centrality(G)
    sort_degree = sorted(degree_list.items(), key=lambda x: x[1], reverse=True)
    degree_char = []
    for i in range (5):
      degree_char.append(sort_degree[i][0])

    closeness_list = nx.closeness_centrality(G)
    sort_closeness = sorted(closeness_list.items(), key=lambda x: x[1], reverse=True)
    closeness_char = []
    for i in range (5):
      closeness_char.append(sort_closeness[i][0])

    betweenness_list = nx.betweenness_centrality(G)
    sort_betweenness = sorted(betweenness_list.items(), key=lambda x: x[1], reverse=True)
    betweenness_char = []
    for i in range (5):
      betweenness_char.append(sort_betweenness[i][0])

    return degree_char,closeness_char,betweenness_char
```
